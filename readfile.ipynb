```python
import pandas as pd

file_path = "F:/KLTN/100_Customer/h100c101.csv"

# Đọc file CSV
df = pd.read_csv(file_path,delimiter=",", header=0)

# Kiểm tra số cột thực tế
print("Số lượng cột thực tế:", df.shape[1])
print("Dữ liệu mẫu:\n", df.head())
```

```python
import numpy as np

# Bỏ hàng đầu tiên (kho hàng) và reset index
customers = df.iloc[1:].reset_index(drop=True)

# Chuyển đổi tọa độ thành numpy array
customer_locations = customers[['x', 'y']].values
depot_location = np.array([df.iloc[0][['x', 'y']].values])  # Kho hàng

print("Toạ độ kho hàng:", depot_location)
print("Toạ độ khách hàng:\n", customer_locations)
```
